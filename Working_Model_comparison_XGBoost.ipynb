```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python

data = pd.read_csv("/Users/hs/AIML/Capstone_Colab/data.csv")
```

```python
data.head()
```

```python
# Remove irrelevant columns
data = data.drop(['step', 'nameOrig', 'nameDest'], axis=1)
```

```python
X = data.drop(['isFraud', 'isFlaggedFraud'], axis=1)
y = data['isFraud']
```

```python
X.head()
```

```python
X.replace(to_replace=['CASH_OUT', 'PAYMENT', 'CASH_IN', 'TRANSFER', 'DEBIT'],value=[1,2,3,4,5],inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python

dt_model=DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_predictions=dt_model.predict(X_test)
```

```python
xgb_model = XGBClassifier()
X_train_xgb = pd.get_dummies(X_train, columns=['type'], drop_first=True)
X_test_xgb = pd.get_dummies(X_test, columns=['type'], drop_first=True)
xgb_model.fit(X_train_xgb, y_train)
xgb_predictions = xgb_model.predict(X_test_xgb)
```

```python
catboost_model = CatBoostClassifier(cat_features=['type'])
catboost_model.fit(X_train, y_train)
catboost_predictions = catboost_model.predict(X_test)
```

```python
xgb_accuracy = accuracy_score(y_test, xgb_predictions)
catboost_accuracy = accuracy_score(y_test, catboost_predictions)
dt_accuracy = accuracy_score(y_test, dt_predictions)
```

```python
xgb_conf_matrix = confusion_matrix(y_test, xgb_predictions)
xgb_precision = precision_score(y_test, xgb_predictions)
xgb_recall = recall_score(y_test, xgb_predictions)
```

```python
catboost_conf_matrix = confusion_matrix(y_test, catboost_predictions)
catboost_precision = precision_score(y_test, catboost_predictions)
catboost_recall = recall_score(y_test, catboost_predictions)
```

```python
dt_conf_matrix = confusion_matrix(y_test, dt_predictions)
dt_precision = precision_score(y_test, dt_predictions)
dt_recall = recall_score(y_test, dt_predictions)
```

```python
xgb_f1_score = f1_score(y_test, xgb_predictions)
catboost_f1_score = f1_score(y_test, catboost_predictions)
dt_f1_score = f1_score(y_test, dt_predictions)
```

```python
print("XGBoost Metrics:")
print(f'Accuracy: {xgb_accuracy}')
print(f'Confusion Matrix:\n{xgb_conf_matrix}')
print(f'Precision: {xgb_precision}')
print(f'Recall: {xgb_recall}')
print(f'F1 Score: {xgb_f1_score}')

print("\n CatBoost Metrics:")
print(f'Accuracy: {catboost_accuracy}')
print(f'Confusion Matrix:\n{catboost_conf_matrix}')
print(f'Precision: {catboost_precision}')
print(f'Recall: {catboost_recall}')
print(f'F1 Score: {catboost_f1_score}')

print("\n Decision Metrics:")
print(f'Accuracy: {dt_accuracy}')
print(f'Confusion Matrix:\n{dt_conf_matrix}')
print(f'Precision: {dt_precision}')
print(f'Recall: {dt_recall}')
print(f'F1 Score: {dt_f1_score}')
```
